```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

# Load Datasets

```python
scheme = pd.read_csv("../Bluestock-Capstone/data/raw/07_scheme_performance.csv")
sip = pd.read_csv("../Bluestock-Capstone/data/raw/04_monthly_sip_inflows.csv")
investors = pd.read_csv("../Bluestock-Capstone/data/raw/08_investor_transactions.csv")
portfolio = pd.read_csv("../Bluestock-Capstone/data/raw/09_portfolio_holdings.csv")

print("Scheme Performance:", scheme.shape)
print("SIP:", sip.shape)
print("Investors:", investors.shape)
print("Portfolio:", portfolio.shape)
```

# Analysis 1: Top Funds by AUM

Assets Under Management (AUM) indicate the size and popularity of a mutual fund.

```python
top_funds = (
    scheme.sort_values("aum_crore", ascending=False)
    .head(10)
)

top_funds[["scheme_name","aum_crore"]]
```

## Observation

The top funds manage a significant proportion of total assets, indicating strong investor preference for established schemes.

# Analysis 2: SIP Growth Trend

```python
sip.head()
```

## Observation

Monthly SIP inflows show consistent long-term growth, reflecting increasing retail investor participation.

# Analysis 3: Investor Demographics

```python
investors["age_group"].value_counts()
```

```python
investors["gender"].value_counts()
```

```python
investors["city_tier"].value_counts()
```

## Observation

# Analysis 4: State-wise Investments

```python
state_investments = (
    investors.groupby("state")["amount_inr"]
    .sum()
    .sort_values(ascending=False)
)

state_investments.head(10)
```

## Observation

A small number of states contribute a large share of total investment activity.

# Analysis 5: Portfolio Sector Exposure

```python
sector_exposure = (
    portfolio.groupby("sector")["weight_pct"]
    .sum()
    .sort_values(ascending=False)
)

sector_exposure
```

## Observation

Sector concentration analysis helps identify diversification opportunities and portfolio risk.

# Key Findings

1. SIP inflows have grown significantly.
2. Mutual fund participation is increasing.
3. A few schemes dominate industry AUM.
4. Investments are concentrated in major states.
5. Portfolio allocation is sector-sensitive.

# Conclusion

The Indian mutual fund industry demonstrates strong growth, increasing investor participation, and healthy long-term investment behavior. These trends create opportunities for asset managers to expand investor reach and product adoption.
